```python
import ast
import pandas as pd
from datasets import load_dataset
```

```python
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
df.sample(5)
```

```python
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])

# Extract day, month, and year
df["job_posted_day"] = df["job_posted_date"].dt.day_name()
df["job_posted_month"] = df["job_posted_date"].dt.strftime("%B")
df["job_posted_month_no"] = df["job_posted_date"].dt.month
df["job_posted_year"] = df["job_posted_date"].dt.year
```

```python
df.sort_values(by=["job_posted_year", "job_posted_month", "job_posted_day"], 
                           ascending=[True, True, True], 
                            inplace=True )

```

```python
median_hour_salary = df.salary_hour_avg.median()

df["salary_hour_avg"] = df["salary_hour_avg"].fillna(median_hour_salary)

df.drop_duplicates(subset=['job_title' , 'company_name'], inplace=True)
```

```python
df.to_csv("cleaned_data.csv", index=False)
```
